```python
#Import dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
stroke_df = pd.read_csv("Resources/healthcare-dataset-stroke-data.csv")
stroke_df
```

```python
#Data Cleaning
# 1. ever_married: Yes = 1, No = 0
stroke_df["ever_married"].replace({"Yes": 1, "No": 0}, inplace=True)
```

```python
# 2. fill in N/A values in bmi column 
# Drop rows with missing values in any of the selected columns
model_df = stroke_df.dropna()

# Convert categorical variables into dummy variables
model_df["gender_Female"] = (model_df["gender"] == "Female").astype(int)
model_df["gender_Male"] = (model_df["gender"] == "Male").astype(int)

# Drop the original "gender" column
model_df = model_df.drop("gender", axis=1)

model_df = pd.get_dummies(model_df, columns=["smoking_status"],drop_first=True)


x = model_df.drop(["bmi", "work_type", "Residence_type"], axis=1)
y = model_df["bmi"]
```

```python
# Create an instance of the linear regression model
model = LinearRegression()

# Fit the model to the data
model.fit(x, y)

# Select rows with missing values in the "bmi" column
missing_data = stroke_df[stroke_df["bmi"].isnull()]
```

```python
# Prepare the missing data for prediction by selecting the same columns as X and creating dummy variables
missing_data["gender_Female"] = (missing_data["gender"] == "Female").astype(int)
missing_data["gender_Male"] = (missing_data["gender"] == "Male").astype(int)
missing_data = missing_data.drop("gender", axis=1)

missing_data = pd.get_dummies(missing_data, columns=["smoking_status"],drop_first=True)
missing_data

# Predict the missing values
predicted_values = model.predict(missing_data.drop(["bmi","work_type", "Residence_type"], axis=1))

# Fill in the missing values in the original DataFrame
stroke_df.loc[stroke_df["bmi"].isnull(), "bmi"] = predicted_values
```

```python
stroke_df.to_csv("Resources/healthcare-dataset-stroke-data-cleaned.csv", index=False)
```
